```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Create a simple line plot
plt.plot(x, y)

# Add labels and title
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Simple Line Plot")

# Display the plot
plt.show()
```

```python
x=[5,8,10]
y=[12,16,6]
plt.plot(x,y)
plt.show()
```

```python
#https://matplotlib.org/stable/gallery/style_sheets/style_sheets_reference.html

from matplotlib import style
style.use('ggplot')
x=[5,8,10]
y=[12,16,6]
plt.plot(x,y)
plt.show()
style.available
```

```python
from matplotlib import style
style.use('ggplot')
x1=[5,8,10]
y1=[12,16,6]
x2=[6,9,11]
y2=[6,15,7]
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Line Plot using style')
plt.plot(x1,y1,'g',label='line one',linewidth=5,color='purple')
plt.plot(x2,y2,'c',label='line two',linewidth=5,color ='orange')
plt.legend(loc=1)
plt.show()


```
