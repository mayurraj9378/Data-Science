```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
ufo=("https://raw.githubusercontent.com/yashy1626/ds_dataset/refs/heads/main/ufo.csv")
ufo=pd.read_csv(ufo)
ufo.head()
```

```python
ufo.shape
```

```python
ufo.info()
```

```python
ufo.isnull().sum()
```

```python
#dropna
#Remove missing values
ufo.dropna().shape
```

```python
#handle missing values
#fillna()
#fill na/NaN values using the specified method
ufo['City'].fillna(value='citymentioned',inplace=True)
ufo['Colors Reported'].fillna(value='citymentioned',inplace=True)
ufo['Shape Reported'].fillna(value='citymentioned',inplace=True)
ufo.isnull().sum()


```
