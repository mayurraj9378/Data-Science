```python
#data visualization - seaborn
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
print(sns.__version__)
```

```python
# built in datasets of seaborn
sns.get_dataset_names()
```

```python
dataset=sns.load_dataset('iris')
dataset
```

```python
#line graph --> lineplot()
sns.lineplot(x='sepal_length',y='sepal_width',data=dataset)
plt.title('Line Graph in seaborn')
plt.show()
```

```python
#customize seaborn plots
sns.set_style('darkgrid',{'grid.color':'k','grid.linestyle':':'})  #linestyle -- : -
sns.lineplot(x='sepal_length',y='sepal_width',data=dataset)
plt.title('Line Graph in seaborn')
plt.show()
```

```python
#change the figure size also
plt.figure(figsize=(10,5))
sns.lineplot(x='sepal_length',y='sepal_width',data=dataset)
plt.title('Line Graph in seaborn')
plt.show()
```

```python
pokemon = pd.read_csv('https://raw.githubusercontent.com/yash240990/Python/master/pokemon.csv')
pokemon
```

```python
#scatter plot in seaborn
import seaborn as sns
import matplotlib.pyplot as plt
sns.scatterplot(x='Attack',y='Defense',hue='Generation',data=pokemon)
plt.show()
```

```python
pokemon.Generation.value_counts().sort_index()
```

```python
pokemon.Generation.value_counts()
```

```python
colors = ['r','y','b','g','c','m']
sns.scatterplot(x='Attack',y='Defense',hue='Generation',data=pokemon,palette=colors)
plt.title('Scatter Plot in seaborn')
plt.show()
```
