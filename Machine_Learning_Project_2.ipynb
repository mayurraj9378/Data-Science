```python
#Supervised Ml -->Polynomial Regreassion
# increase  the accuracy
import pandas as pd
import numpy as np

import matplotlib as plt
import matplotlib.pyplot as plt
import sklearn.linear_model as lm
lin_reg = lm.LinearRegression()


```

```python
data=pd.read_csv('https://raw.githubusercontent.com/yash240990/Python/master/Position_Salaries.csv')
data.head()
```

```python
x= data.Level.values[:,np.newaxis]
x
```

```python
y=data.Salary.values
# y dependent variable
y
```

```python
#buliding model using simple liear regression
lin_reg.fit(x,y)
```

```python
pred_values = lin_reg.predict(x)
pred_values
```

```python
#plot actual values
plt.plot(x,y,color='red',label='Actual Values')
plt.scatter(x,y,color='red')
#plot predicted values
plt.plot(x,pred_values,color='green',label='Predicted values')
plt.scatter(x,pred_values,color='red')
plt.title('Detection  modelusing simple linear regression Degree=2')
plt.xlabel('Positional level')
plt.ylabel('Salaries')


plt.legend()
plt.show()
```

```python
#Building a model using Polynomial Regression
#In Polynomial regression we increase the degree of independent
#variable to achieve more accuracy

from sklearn.preprocessing import PolynomialFeatures
x
```

```python
x.shape
```

```python
poly_reg=PolynomialFeatures(degree=2)
x_poly_2=poly_reg.fit_transform(x)
x_poly_2

#when degree is set to 2 , new feature will be created [1 x*1 x*x]
# when degree is set to 3 , new feature will be created [1 x*1 x*x x*x*x]
# when degree is set to 4 , new feature will be created [1 x*1 x*x x*x*x x*x*x*x]
```

```python
lin_reg.fit(x_poly_2,y)
```

```python
pred_values = lin_reg.predict(x_poly_2)
pred_values
```

```python
poly_reg=PolynomialFeatures(degree=3)
x_poly_3=poly_reg.fit_transform(x)
x_poly_3
pred_values = lin_reg.predict(x_poly_3)
pred_values
```

```python
lin_reg.fit(x_poly_3,y)
```

```python
#plot actual values
plt.plot(x,y,color='red',label='Actual Values')
plt.scatter(x,y,color='red')
#plot predicted values
plt.plot(x,pred_values,color='green',label='Predicted values')
plt.scatter(x,pred_values,color='red')
plt.title('Detection  modelusing simple linear regression Degree=3')
plt.xlabel('Positional level')
plt.ylabel('Salaries')


plt.legend()
plt.show()
```
