```python
# Supervised Machine Learning
# Regression : Multivariate / Multiple Linear Regression
# STAGE 1 : Data Gathering / Load the Dataset

import pandas as pd
data = pd.read_csv('https://raw.githubusercontent.com/yash240990/Python/master/Startups_Data.csv')
data.head()
```

```python
# STAGE 2 : EDA
data.shape
```

```python
data.columns
```

```python
data.info()
```

```python
# STAGE 3 : DATA PREPERATION
# a.) Check for missing and duplicate values
# b.) Convert non-numerical features into numerical features
# c.) Perform normalization
# d.) Select dependent and independent variable
```

```python
# a.) Check for missing and duplicate values
print('Missing values in each column:')
print(data.isnull().sum())
```

```python
print(f'Duplicate rows: {data.duplicated().sum()}')
```

```python
# b.) Convert non-numerical features into numerical features
#LabeEncoder   OneHotEncoder     ColumnarTransformer
# California --> 0
# Florida --> 1
# Newyork --> 2

from sklearn.preprocessing import LabelEncoder , OneHotEncoder
from sklearn.compose import ColumnTransformer
x = data.iloc[:,:-1].values
x
label_encoder = LabelEncoder()
x[:,3] = label_encoder.fit_transform(x[:,3])
transformer = ColumnTransformer(transformers=[("OneHot",OneHotEncoder(),[3])],
                                remainder='passthrough')
# remainder = 'passthrough' means do not apply anything on remaining column
x = transformer.fit_transform(x)
x = x.astype(int)
```

```python
x
# california   florida   newyork   r_d     admin    market
#  0             0          1      165349   136897  471784
#  1             0          0      162597  151377   443898
```

```python
# d.) Select dependent and independent variable
x = data[['R&D Spend', 'Administration', 'Marketing Spend', 'State']]
x = x.values
x
```

```python
y = data.Profit.values
y
```

```python
# Apply one-hot encoding to the 'State' column
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore')
# Access the 'State' column using its index (3 in this case)
encoded_state = encoder.fit_transform(x[:, [3]]) # Use x[:, [3]] to select the 'State' column
encoded_df = pd.DataFrame(encoded_state, columns=encoder.get_feature_names_out(['State']))
# Convert x back to DataFrame for using column name in drop
x = pd.DataFrame(x) # Convert x to a pandas DataFrame for using column name in drop
x = x.drop(3, axis=1)  # Drop the original 'State' column using index
x = pd.concat([x, encoded_df], axis=1)
x = x.values # Convert back to NumPy array if needed
y = data.Profit.values
```

```python
# Split The Dataset Into Training And Testing Subset
from sklearn.model_selection import train_test_split
# test_size=0.2 means 20% of the data will be used for Testing and remaining 80% used to training
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
print('Original Shape')
print(x.shape)
print(y.shape)
print('After Splitting')
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
# Model Development
from sklearn.linear_model import LinearRegression
mlr = LinearRegression()
mlr.fit(x_train,y_train)
```

```python
pred_values = mlr.predict(x_test)
pred_values
```

```python
from sklearn.metrics import r2_score
accuracy = r2_score(y_test,pred_values) #y_test=original value pred_values:predicted value
print('Accuracy of Model :',accuracy)
print('Accuracy of Model :',accuracy*100,'%')
```
